```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self):
        return "Hace un sonido"


class Gato(Animal):
    def hablar(self):
        return "Miau miau!"


Gato = Gato("gatita")
print(f"{Gato.nombre} dice: {Gato.hablar()}")
```

```python
class Vehiculo:
    def encender(self):
        print("El vehículo está encendido")


class Automovil(Vehiculo):
    def abrir_puertas(self):
        print("Las puertas están abiertas")


auto = Automovil()
auto.encender()
auto.abrir_puertas()
```

```python
class SerVivo:
    def respirar(self):
        print("Está respirando")


class Animal(SerVivo):
    def moverse(self):
        print("Se está moviendo")


class Ave(Animal):
    def volar(self):
        print("Está volando")


paloma = Ave()
paloma.respirar()
paloma.moverse()
paloma.volar()
```

```python
class Figura:
    def dibujar(self):
        print("Dibujando figura genérica")


class Cuadro(Figura):
    def dibujar(self):
        print("Dibujando un cuadro")


class Circulo(Figura):
    def dibujar(self):
        print("Dibujando un círculo")


figuras = [Cuadro(), Circulo()]

for figura in figuras:
    figura.dibujar()
```

```python
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre


class Empleado(Persona):
    def __init__(self, nombre, cargo):
        super().__init__(nombre)
        self.cargo = cargo


emp = Empleado("Carlos", "Desarrollador")
print(f"{emp.nombre} - {emp.cargo}")
```

```python
import math

class Figura:
    def area(self):
        return 0


class Cuadro(Figura):
    def __init__(self, lado):
        self.lado = lado

    def area(self):
        return self.lado ** 2


class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2


figuras = [Figura(), Cuadro(4), Circulo(3)]

for f in figuras:
    print(f"Área: {f.area()}")
```
